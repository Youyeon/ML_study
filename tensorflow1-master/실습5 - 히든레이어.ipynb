```python
# 라이브러리 사용
import pandas as pd
import tensorflow as tf
```

# 보스턴 집값 예측

```python
# 과거의 데이터를 준비합니다.
path = 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/boston.csv'
boston = pd.read_csv(path)
boston
```

```python
# 종속변수, 독립변수
ind_val = boston[['crim','zn','indus','chas','nox','rm','age','dis','rad','tax','ptratio','b','lstat']]
dep_val = boston[['medv']]
print(ind_val.shape, dep_val.shape)
```

```python
# 모델의 구조를 만듭니다.
X = tf.keras.layers.Input(shape=[13])
H = tf.keras.layers.Dense(10, activation='swish')(X) #10개의 노드, 활성화함수 swish
Y = tf.keras.layers.Dense(1)(H)

model = tf.keras.models.Model(X, Y)
model.compile(loss='mse')
```

```python
# 모델의 구조 확인
model.summary()
```

```python
# 데이터로 모델을 학습(FIT)합니다.
model.fit(ind_val, dep_val, epochs=1000, verbose=0)
model.fit(ind_val, dep_val, epochs=10)
```

```python
# 모델을 이용합니다.
print(model.predict(ind_val[:5]))
print(dep_val[:5])
```

# 아이리스 품종 분류

```python
# 과거의 데이터를 준비합니다.
path = 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/iris.csv'
iris = pd.read_csv(path)
# 원-핫 인코딩
iris = pd.get_dummies(iris)
iris
```

```python
# 종속변수, 독립변수
ind_val = iris[['꽃잎길이', '꽃잎폭', '꽃받침길이', '꽃받침폭']]
dep_val = iris[['품종_setosa', '품종_versicolor', '품종_virginica']]
print(ind_val.shape, dep_val.shape)
```

```python
# 모델의 구조를 만듭니다.
X = tf.keras.layers.Input(shape=[4])
H = tf.keras.layers.Dense(8, activation='swish')(X)
H = tf.keras.layers.Dense(8, activation='swish')(H)
H = tf.keras.layers.Dense(8, activation='swish')(H)
Y = tf.keras.layers.Dense(3, activation='softmax')(H)

model = tf.keras.models.Model(X, Y)
model.compile(loss='categorical_crossentropy', metrics='accuracy')
```

```python
# 모델의 구조 확인
model.summary()
```

```python
# 데이터로 모델을 학습(FIT)합니다.
model.fit(ind_val, dep_val, epochs=100, verbose=0)
```

```python
# 모델을 이용합니다.
print(model.predict(ind_val[0:5]))
print(dep_val[0:5])
```
